```python
from sqlalchemy import create_engine
import pandas as pd
from pathlib import Path

ROOT   = Path.cwd().parent
engine = create_engine(f'sqlite:///{ROOT}/data/processed/hospital.db')

df = pd.read_csv(ROOT / 'data' / 'processed' / 'hospital_clean.csv')
df.to_sql('encounters', engine, if_exists='replace', index=False)

print(f'Loaded {len(df):,} rows into hospital.db')
```
